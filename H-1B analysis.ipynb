```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import xlrd
```

```python
df = pd.read_csv('H1B2019.csv')
```

# load data 

```python
df=df[df.VISA_CLASS=='H-1B']
```

# data cleaning

```python
#crate state list
state_list = list(df.EMPLOYER_STATE.unique())
state_list.remove(np.nan)

#clean job title & employer name
df['JOB_TITLE']=[i[:-1] if str(i)[-1]=='.' else i for i in df['JOB_TITLE']]
df['EMPLOYER_NAME']=[i[:-1] if str(i)[-1]=='.' else i for i in df['EMPLOYER_NAME']]
df['EMPLOYER_NAME']=[i[:i.index('  ')]+i[i.index('  ')+1:] if '  ' in str(i) else i for i in df['EMPLOYER_NAME']]
```

# filter job title

```python
# the title of job includes analyst
df_=df['JOB_TITLE'].unique()
data_job_title=[i for i in df_ if 'ANALYST'  in str(i) ]
data_job_title=[i for i in data_job_title if 'SENIOR' not in str(i) ]
data_job_title=[i for i in data_job_title if 'SR.' not in str(i) ]
data_job_title=[i for i in data_job_title if 'SR' not in str(i) ]

data_job_title=list(set(data_job_title))
data_job_title
```

# data visualization

```python
# state
data_geo=df[df.JOB_TITLE.isin(data_job_title)].groupby('EMPLOYER_STATE')['CASE_NUMBER'].count().sort_values(ascending=False).to_frame()
data_geo.to_csv('./data_geo.csv')
```

```python
# analyst in every company
data_company=df[df.JOB_TITLE.isin(data_job_title)].groupby('EMPLOYER_NAME')['CASE_NUMBER'].count().sort_values(ascending=False).to_frame()
data_company=data_company[data_company.CASE_NUMBER>=20].reset_index()
data_company=data_company.groupby('EMPLOYER_NAME')['CASE_NUMBER'].sum().sort_values(ascending=False).to_frame()
data_company.to_csv('./data_company.csv')
data_company
```

```python
# analyst in IL
data_comp=df[(df.EMPLOYER_STATE=='IL') & (df.JOB_TITLE.isin(data_job_title))].groupby('EMPLOYER_NAME')['CASE_NUMBER'].count().sort_values(ascending=False).reset_index()
data_comp_il=pd.merge(data_comp,df[['EMPLOYER_CITY','EMPLOYER_STATE','EMPLOYER_NAME']].drop_duplicates(),on=['EMPLOYER_NAME'],how='left')
data_comp_il.to_csv('data_comyil.csv')
```

```python
data_title_times=df[(df.EMPLOYER_STATE=='IL') & (df.JOB_TITLE.isin(data_job_title))].groupby('JOB_TITLE')['CASE_NUMBER'].count().to_frame()
data_title_times.to_csv('job_IL.csv')
```
